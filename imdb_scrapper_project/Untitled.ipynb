```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen


def amazon(query):
    query=query.replace("",'+')
    link=f"https://www.amazon.in/s?k={query}&crid=2TO4BCBIL21VJ&sprefix=philips+hair+drye%2Caps%2C420&ref=nb_sb_noss_1"
        
        
    try:
        req=requests.get(link).text
        clean_html=bs(req,'html.parser')
        container=clean_html.find_all("div",{"class":"sg-row"})
        product_name=container[1].find('span',{"class":"a-size-medium a-color-base a-text-normal"}).text
        product_price=container[1].find("span",{"class":"a-offscreen"}).text
        product_link="https://www.amazon.in/"+container[1].a['href']
        product_image=container[1].img['src']
        product_rating=container[1].find('span',{"class":'a-size-base'}).text
        product_total_rating=container[1].find("span",{"class":"a-size-base s-underline-text"}).text
        product_total_rating=product_total_rating.replace('(','')
        product_total_rating=product_total_rating.replace(')','')



    except Exception:
        req=requests.get(link).text
        clean_html=bs(req,'html.parser')
        container=clean_html.find('div',{"class":"s-main-slot s-result-list s-search-results sg-row"})
        c=container.find_all('div',{"class":"s-widget-container s-spacing-small s-widget-container-height-small celwidget slot=MAIN template=SEARCH_RESULTS widgetId=search-results_1"})


        product_name=c[0].find('span',{"class":"a-size-base-plus a-color-base a-text-normal"}).text
        product_price=c[0].find('span',{"class":"a-price-symbol"}).text+c[0].find('span',{"class":"a-price-whole"}).text
        product_link="https://www.amazon.in/"+c[0].a['href']
        product_image=c[0].img['src']
        product_rating=c[0].find('span',{"class":"a-size-base"}).text
        product_total_rating=c[0].find('span',{"class":"a-size-base s-underline-text"}).text
        product_total_rating=product_total_rating.replace('(','')
        product_total_rating=product_total_rating.replace(')','')

    return product_name,product_price,product_link,product_image,product_rating,product_total_rating
```
